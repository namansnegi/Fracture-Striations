```python
import math

def grayscale(image):    	
        
    return cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
                
def gaussian_blur(img, kernel_size):
    return cv2.GaussianBlur(img,(kernel_size, kernel_size),0)
    
def auto_canny(image, sigma=0.25):
    #compute the median of the single channel pixel intensities
    v = np.median(image) #apply automatic Canny edge detection using the computed median
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    edged = cv2.Canny(image, lower, upper)
    return edged
 
def region_of_interest(img, vertices):
    # Next we'll create a masked edges image using cv2.fillPoly()
    mask = np.zeros_like(img)   
        #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255  
    
    # This time we are defining a four sided polygon to mask
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    
    return masked_image
    
def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    # Iterate over the output "lines" and draw lines on a blank image
    for line in lines:
        
        for x1,y1,x2,y2 in line:
            cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)
       
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]),min_line_len, max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )    
    
def line_selection(lines):
 
    final_lines=[]
    c=0  
    for x1,y1,x2,y2 in lines[:, 0]:
        
        m = (float(y2) - y1) / (x2 - x1)
        if m>0:
            if (m<0.25 and m>0.05):
                final_lines.append(lines[c,:])
        if m<0:
            if (m>-0.25 and m<-0.05):
                final_lines.append(lines[c,:])
        c=c+1
   
    return final_lines
```

```python
import os
os.listdir("samples/")



# Main Function
# 1. Define region of interest
# 2. Run Canny Edge Detection on grayscale image passed through a gaussian filter
# 3. Mask the region of interest based on found edges and defined vertices 
# 4. Find Hough Lines in the masked region
# 5. Draw lines to higlight the Hough lines and overlay the image on the origianl unprocessed image
 
def pipeline(image):  
    ### Params for region of interest
    bot_left = [50, 540]
    bot_right = [980, 545]
    apex_right = [515, 320]
    apex_left = [450, 315]
    vertices = [np.array([bot_left, bot_right, apex_right, apex_left], dtype=np.int32)]
    if len(image.shape)==3:
        ### Run canny edge detection and mask region of interest
        gray = grayscale(image)
    else:
        gray = image
    
    
    blur = gaussian_blur(gray, 3)
    
    edge = feature.canny(blur,sigma=0.5)
    
    
    ### Run Hough Lines and separate by +/- slope
    
    lines = cv2.HoughLinesP(edge, 1, np.pi/180, 10, np.array([]), 5,10)

    final_lines=line_selection(lines)
    
    ### Draw lines and return final image 
    line_img = np.copy((image)*0)
    draw_lines(line_img,final_lines, thickness=1)
    #draw_lines(line_img,lines, thickness=1)
    
    #line_img = region_of_interest(line_img, vertices)
    final_image = weighted_img(line_img, image)
    
    return final_image
```

```python
# Function to run the pipeline on imgaes or video    
def process_image(image):
    result = pipeline(image)
    return result

image = plt.imread('samples/CET01_08_rotated_z1.tif')
image=image[:,:,0]
image = 255.0*(image/255.0)**2
output=process_image(image,)
plt.imshow(output,cmap=plt.cm.gray)
```
